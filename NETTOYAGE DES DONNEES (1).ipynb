```python
#Supprimer des colonnes inutiles
df.drop(["SibSp", "Parch", "Ticket", "Cabin", "Embarked"], axis=1, inplace=True)

#df.drop(...) : Supprimer quelque chose du DataFrame
#axis=1 La liste des colonnes √† supprimer
#inplace=True :La suppression est faite directement dans df
df.columns






```

```python
#D√©finir PassengerId comme index
df.set_index("PassengerId", inplace=True)
```

```python
#Afficher les valeurs manquantes par colonne
df.isnull().sum()
```

```python
#Remplir les √¢ges manquants par la m√©diane selon la classe
df['Age'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
#Cr√©er la colonne Title √† partir de Name
df['Title'] = df['Name'].str.extract('([A-Za-z]+)\.')
```

```python
#Compter les titres
df['Title'].value_counts()
```

```python
#Convertir les titres en chiffres
df['Title'] = df['Title'].map({'Mr':0, 'Miss':1, 'Mrs':2}).fillna(3)
df
```

```python
#Convertir la colonne Sex
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
df
```

```python
#Transformer la colonne Fare avec loc
df.loc[df['Fare'] <= 17, 'Fare'] = 0
df.loc[(df['Fare'] > 17) & (df['Fare'] <= 30), 'Fare'] = 1
df.loc[(df['Fare'] > 30) & (df['Fare'] <= 100), 'Fare'] = 2
df.loc[df['Fare'] > 100, 'Fare'] = 3
```
