Ce notebook présente des exercices pratiques pour apprendre NumPy :

```python
import numpy as np
```

```python
#creation dun tableau x contennat ds nbr de 10 a 50
x = np.arange(10,51)
```

```python
#creation d un table 3*3 remplit de nbr aleatoir entre 10 et 50
#randint nbr aleatoir entre 10 et 50
#size(3,3) 3*3
x = np.random.randint(10,51,size=(3,3))
```

```python
#remplacer tous ls nbrs paires par 0
#%2==0 pour detecter les nbrs paires
x[x%2==0] = 0
print(x)
```

```python
#afficher lelemnt situer a la deuxieme ligne colonne 3
#2 eme ligne index 1  && 3 eme collonne indice 2
print(x[1, 2])
```

```python
#slicing
#extraire premiere ligne
a = x[0,:]
#extraire derniere colonne 
b = x[:,-1]
print(a,b)
```

```python
#fancy indexing
#extraire les deux premier ligne et les deux premiere collonnes
x[:2,:2]
#:2 pour ligne :  commence de  ligne 0 2 arrete avant ligne 2 donc lign 0 et 1 deux premiere ligne
#:2 pour colonne ; commence de collone 0 arrete avant colonne 2 donc collonne 0 1 les deux premiere colonne

```

```python
#fancy indexing
#extraire ls elements:1ere et 3 eme ligne ;1 ere et 3 eme colonne
x[[0,2],:] [:,[0,2]]

    
```

```python
#boolean indexing
#afficher tous ls elemnts superieur a 30
x[x>30]
```

```python
#boolean indexing
#remplacer tous ls elemtns inferieur a 20 par 20
x[x<20] = 20
print(x)
```

```python
#statistiques
#calculer la somme de chaque ligne et chaque collonne
#axis=1 veut dire horizzental veut dire ligne 
#sxis=0 cad vertical cas colonne
xl= x.sum(axis=1)
xc= x.sum(axis=0)
print(xl,xc)
#axis=1 ligne
#axis=0 colonne
```

```python
#statistique
#calculer min max moyenne de tableau
print(x.sum())
print(x.min())
print(x.max())
print(x.mean())
```

```python
#statistiques
#calculer moyenn,min,max de chaque collonne et chaaque ligne

x.min(axis=1) , x.max(axis=1) , x.mean(axis=1)
#calculer moyenn,min,max de chaque collone
x.max(axis=0) , x.min(axis=0) , x.mean(axis=0)



```

```python
#normalisation
#normaliser le tableau
#x.mean() pour calculler moyennn ;x.std calcul d ecart type et x-x.mean()/x.std pour centrer et reduit les valeur
#x-x.mean() la soustraction de moyenne cad le centre devien 0 ;  on devise par lecart type la variance devien 1 
x_normalise = (x-x.mean() /x.std())
```
