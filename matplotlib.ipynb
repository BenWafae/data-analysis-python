
# TP NumPy & Matplotlib – Visualisation de données

Ce notebook présente des exercices pratiques pour apprendre à :
- Manipuler des tableaux avec NumPy
- Créer des graphiques avec Matplotlib
- Visualiser des données numériques sous différentes formes
- Tracer des courbes, histogrammes, tableaux 2D et scatter plots



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x,y)
plt.title("lignnee simplee")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
#plt.plot(x, y) trace une ligne reliant les points
#plt.title, plt.xlabel, plt.ylabel ajoutent des informations
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 3, 9, 1, 7]
#scatter veut dire diagramme de point,montre chaque point individuellement
plt.scatter(x,y,color='red')
plt.title("diagramme de point")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#graphique en barre
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]
plt.bar(categories,values,color='blue')
plt.title("Graphique en barres")
plt.show()
#plt.bar() → crée un graphique en barres verticales utile pour comparer des valeurs entre catégories
```

```python
table= np.arange(1,26).reshape(5,5)
#affichage d un table 2d comme une image 
#cmap pallette de couleur sympa
plt.imshow(table, cmap = 'viridis')
#barre de couleur lechelle de valuer
plt.colorbar()
plt.title("table 2d")
```

```python
x=np.linspace(0,4,100)
print(x)
y= 2*np.sin(x)
#reacer y en fct de x
plt.plot(x,y,label='y=2*sin(x)',color='blue')
plt.title("fonction sinus")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
x = np.linspace(0,2*np.pi,100)
#np.sinus calcule sinus pour chaque eleement de tableau
y1=np.sin(x)
y2=0.5 * np.cos(2*x)
print(y1[:10])
print(y2[:10])
#tracer les deux fct
plt.plot(x, y1, label='y1 = sin(x)', color='blue')
plt.plot(x, y2, label='y2 = 0.5*cos(2x)', color='red', linestyle='--')
plt.title("Fonctions sinus et cosinus")
plt.xlabel("x")
plt.ylabel("y")
#plt.legend() → montre quelle courbe correspond à quelle fonction
plt.legend()
plt.grid(True)
```

```python
x = np.random.rand(20) * 10  # valeurs entre 0 et 10
y = np.random.rand(20) * 5   # valeurs entre 0 et 5
#plt scatter trace chaque pnntt x[i], y[i]
plt.scatter(x, y, color='green', s=100, marker='o', alpha=0.7)
plt.title("Diagramme de points")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
