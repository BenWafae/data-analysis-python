```python
#creation dune serie dans lequel ls index sont ls jrs de semain et ls temperatures des valeurs
temp = pd.Series([22,24,19,23,25,21,20],index=["lundi","Mardi","Mercredi","Jeudi","Vendredi","Samedi","Dimanche"])
print(temp)
```

```python
#courbe de levolution des temperatures au cours de la semaine 
#temp.index :les jrs axe X
#temp.values: les temperatures axe Y
plt.plot(temp.index, temp.values,marker = "o")
plt.title("volution des tempratures au cours de la semaine")
plt.xlabel("Jours de la semaine")
plt.ylabel("Temprature (C)")
plt.grid()
plt.show()
```

```python
#creation d un dataframe df pour representerces nfos les noms de colonnes sont ville.temp,humd,precip
#ici on a creer un dictionnaire dabord pour le remplir par des infos et puis on va utiliser dataframe pour les organiser
dic={"ville":["casablanca","Marakech","Fes","Agadir","Rabat"],
     "Temp":[18.5,20.0,19.0,19.5,18.0],
     "Humd":[72,50,60,65,70],
     "Precip":[430,250,400,300,450]}
dic
```

```python
df = pd.DataFrame(dic)
df
#df pour laffichage 
```

```python
# Afficher la colonne des villes
df["ville"]
df.ville
#les deux sont correct 
```

```python
#ligne correspondat a marakesh
df[df["ville"]=="Marakech"]
```

```python
#ville avec temp>19
df[df["Temp"]>19]
```

```python
#ville avec le plus de preception
df.loc[df["Precip"].idxmax(),"ville"]
#df["Precip"] pour selectionne la colonne precip
#loc sert  slectionner des lignes et colonnes par label ou index.
#ville pour afficher le nom de ville ps just lindex
```

```python
#tracer la courbe
df["Temp"].plot(marker='o',color='b',label="Temp")
df["Humd"].plot(marker='x',color='r',label="Humd")
df["Precip"].plot(marker='+',color='g',label="Precip")
plt.title("Meteo")
plt.xlabel("ville")
plt.ylabel("Temperature C")
plt.grid(True)
plt.show()
#ici on a utiliser just dataframe df au lieu dutiliser matplotlib 
#.plot integrer a pandas et utilise matplotlib en arriere plan
```
